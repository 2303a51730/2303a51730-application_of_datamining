```python
df['Glucose_Level'].fillna(df['Glucose_Level'].mean(), inplace=True)
```

```python
df['Glucose_Level'] = df['Glucose_Level'].fillna(df['Glucose_Level'].mean())

```

```python
df['Blood_Pressure'] =df['Blood_Pressure'].fillna(df['Blood_Pressure'].mean())
```

```python
df['Insulin'] = df['Insulin'].fillna(df['Insulin'].mean())
```

```python
df
```

```python
df['BMI'].fillna(df['BMI'].mean(),inplace=True)
```

```python
print("Dataset after handling missing values:")
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Glucose_Level']] = scaler.fit_transform(df[['Age', 'Glucose_Level']])
```

```python
df['Blood_Pressure'] = (df['Blood_Pressure'] - df['Blood_Pressure'].min()) / (df['Blood_Pressure'].max() - df['Blood_Pressure'].min())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler and assign it to s_scaler
s_scaler = StandardScaler()

df[['BMI']] = s_scaler.fit_transform(df[['BMI']])
```

```python
df['Insulin'] = (df['Insulin'] - df['Insulin'].mean()) / df['Insulin'].std()
```
